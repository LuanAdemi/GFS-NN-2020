```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math
import numpy as np
%matplotlib inline
from matplotlib import pyplot
import matplotlib.pyplot as plt
from IPython.core.debugger import set_trace
from IPython.display import clear_output
```

```python
device = torch.device("cuda")
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()

        self.l1 = nn.Linear(1, 32)
        self.l2 = nn.Linear(32, 32)
        self.l3 = nn.Linear(32, 1)

    def forward(self, state):
        a = F.relu(self.l1(state))
        a = F.relu(self.l2(a))
        return torch.tanh(self.l3(a))
```

```python
data_set = []
label_set = []

for i in range(80):
    label_set.append([math.sin(i/4)])
    data_set.append([i/4])
    
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1, 1))

data_set = scaler.fit_transform(data_set)
label_set = scaler.fit_transform(label_set)

data = torch.FloatTensor(data_set).to(device)
labels = torch.FloatTensor(label_set).to(device)
```

```python
model = Net().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 500
loss_holder = []
for i in range(epochs):
    for j in range(80):
        
        y_pred = model(data[j])
        single_loss = F.mse_loss(y_pred, labels[j])
        
        optimizer.zero_grad()
        single_loss.backward()
        optimizer.step()
        
    if i%25 == 1:
        loss_holder.append(single_loss.item())
        clear_output()
        plt.title('Loss')
        plt.ylabel('y')
        plt.xlabel('x')
        plt.grid(True)
        plt.autoscale(axis='x',tight=True)
        plt.plot(loss_holder)
```

```python
model.eval()

approx_data = []

for i in range(80):
    approx_data.append(model(data[i]).item())
    #set_trace()
    #print(approx_data)
real_data = scaler.inverse_transform(np.array(approx_data).reshape(-1, 1)) 

pyplot.scatter(data_set,label_set, label='Actual')
# plot x vs yhat
pyplot.scatter(data_set,approx_data, label='Predicted')
pyplot.title('Input (x) versus Output (y)')
pyplot.xlabel('Input Variable (x)')
pyplot.ylabel('Output Variable (y)')
pyplot.legend()
pyplot.show()
```
