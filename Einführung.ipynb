# Einführung: Was sind neuronale Netze

Neuronale Netze sind einer der wichtigsten Elemente in der modernen Forschung rund um das maschinelles Lernen. Sie werden als Funktionsapproximatoren verwendet um durch einen Lernalgorithmus daruf trainiert zu werden, $X$ einem Wert $Y$ zuzuweisen.

Sie bestehen meistens aus einfachen Neuronen, welche $n$ Eingaben $X_n$ und eine Ausgaben $Y$ besitzen.
<br><br>
Wobei $Y$ durch $Y=f(\sum_{i=1}^{n} W_i * X_i + B_i)$ definiert ist.

Jedoch ist der Nutzen so eines Neurones alleine sehr minimal. Deswegen werden diese in riesigen Netzwerken verbaut.

Jedes dieser Neuronen besitzt eine Aktivierungsfunktion. Im Laufe der Jahre wurden immer wieder neue Aktivierungsfunktionen eingeführt. Hier sind mal zwei weit verbreitete:

- Eine sehr häufig verwendete Aktivierungsfunktion ist der **tangens hyperbolicus** $\tanh(x)$:

```python
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.functional as F

x = np.arange(-5, 5, 0.1)

plt.title("Tanh")
plt.grid(True)
plt.xlabel('x', fontsize=15)
plt.ylabel('f(x)', fontsize=15)
plt.plot(np.tanh(x))
```

- Des weiteren gibt es noch die ReLU Funktion (Rectified Linear Unit)

```python
plt.grid(True)
plt.xlabel('x', fontsize=15)
plt.ylabel('f(x)', fontsize=15)
plt.plot(np.maximum(0, x))
```

Nun wollen wir von einem Datensatz mit $(x,y)$ paaren unser Netzwerk so trainieren, dass es uns bei der Eingabe $x$ die Ausgabe $y$ liefert.
Dafür müssen wir irgendwie $W$ und $B$ so anpassen, dass dies der Fall ist.
